```python
import psycopg2
import pandas as pd

# 1. Load DataFrame
df = pd.read_csv("cleaned_sheet.csv")
print("Worksheet loaded âœ…")

# Optional: if you don't want the auto-index column from CSV, uncomment:
# df = df.loc[:, ~df.columns.str.startswith("Unnamed")]

# 2. Connect to PostgreSQL
conn = psycopg2.connect(
    host="localhost",
    database="mospi",
    user="postgres",
    password="12345",
    port="5432"
)
cur = conn.cursor()
print("Connected to PostgreSQL âœ…")

# 3. Fixed table name
table_name = "inflation"

# 4. Drop the table if it exists
cur.execute(f"DROP TABLE IF EXISTS {table_name}")
conn.commit()

# 5. Map Pandas dtypes to PostgreSQL
dtype_mapping = {
    "int64": "BIGINT",
    "float64": "DOUBLE PRECISION",
    "object": "TEXT"
}

columns = []
print("\nðŸ“Œ Column â†’ PostgreSQL type mapping:")
for col, dtype in df.dtypes.items():
    pg_type = dtype_mapping.get(str(dtype), "TEXT")  # fallback: TEXT
    print(f" - {col}: {dtype}  â†’  {pg_type}")
    columns.append(f'"{col}" {pg_type}')

# Create table with mapped types
create_table_query = f"CREATE TABLE {table_name} ({', '.join(columns)});"
cur.execute(create_table_query)
conn.commit()
print(f"\nTable '{table_name}' created âœ…")

# 6. Insert DataFrame data into PostgreSQL (row by row)
num_cols = df.shape[1]
placeholders = ', '.join(['%s'] * num_cols)

# Build column names string, then ESCAPE any literal '%' by doubling them (%%)
col_names = ', '.join([f'"{col}"' for col in df.columns])
safe_col_names = col_names.replace('%', '%%')   # <-- important: escape % in identifiers/text

insert_query = f'INSERT INTO {table_name} ({safe_col_names}) VALUES ({placeholders})'
# Debug print (uncomment if you want to inspect the final query string)
# print("DEBUG INSERT QUERY:", insert_query)

for i, row in enumerate(df.itertuples(index=False, name=None), start=1):
    # optional debug/check:
    # if len(row) != num_cols:
    #     raise RuntimeError(f"Row length {len(row)} != num_cols {num_cols} (row index {i})")

    cur.execute(insert_query, row)

conn.commit()
print("Data inserted into PostgreSQL âœ…")

# 7. Verify the data
cur.execute(f"SELECT * FROM {table_name} LIMIT 5;")
rows = cur.fetchall()
print("\nSample data from table:")
for r in rows:
    print(r)

# 8. Close connection
cur.close()
conn.close()
print("\nPostgreSQL connection closed âœ…")
```
